```python
# Import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
import prettytable as pt
```

```python
# Data Collection

read_file = pd.read_csv (r'data/transfusion.data')
read_file.to_csv (r'transfusion.csv', index=None)
df = pd.read_csv('transfusion.csv')
df.head()
```

```python
# Preprocessing

# define the imputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# transform the dataset
val = imputer.fit_transform(df.values)
ds = pd.DataFrame(val)
ds.columns = ["Recency (months)", "Frequency (times)", "Monetary (c.c. blood)", "Time (months)", "whether he/she donated blood in March 2007"]

not_donate = ds.loc[ds['whether he/she donated blood in March 2007'] == 1]
ds = pd.concat([ds, not_donate])
ds = pd.concat([ds, not_donate])
ds['whether he/she donated blood in March 2007'].value_counts()
```

```python
# Show the data distribution by average, standard deviation, and value frequency(binning might be needed) of every feature.
feature_index = ["Recency (months)", "Frequency (times)", "Monetary (c.c. blood)", "Time (months)", "whether he/she donated blood in March 2007"]
ds_label1 = ds.loc[ds["whether he/she donated blood in March 2007"] == 0]
ds_label2 = ds.loc[ds["whether he/she donated blood in March 2007"] == 1]
ds_split = [ds_label1, ds_label2]
label = ['not donate', 'donate']
width = [4, 2, 500, 5]

for feature_num in range(0, len(feature_index)-1):
    plt.figure(figsize = (10, 7))
    count_0 = ds_split[0][feature_index[feature_num]]
    count_1 = ds_split[1][feature_index[feature_num]]
    
    count_0 = list(count_0)
    count_0 = [x / width[feature_num] for x in count_0]
    count_0 = [int(x) for x in count_0]
    
    count_1 = list(count_1)
    count_1 = [x / width[feature_num] for x in count_1]
    count_1 = [int(x) for x in count_1]
    
    x_0 = []
    y_0 = []
    for i in range(0, np.max(count_0)+1):
        x_0.append(count_0.count(i))
        y_0.append(i)
        
    x_1 = []
    y_1 = []
    for i in range(0, np.max(count_1)+1):
        x_1.append(count_1.count(i))
        y_1.append(i)

    plt.plot(y_0, x_0, label='not donate')
    plt.plot(y_1, x_1, label='donate')
    plt.xlabel(feature_index[feature_num])
    plt.ylabel("value frequency")
    plt.legend(loc=1, prop={'size': 12})
    plt.show()
```

```python
# X is data without column 'class'
X = ds[["Recency (months)", "Frequency (times)", "Monetary (c.c. blood)", "Time (months)"]].values
# y is data only contain column 'class'
y = ds['whether he/she donated blood in March 2007'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=100)
```

```python
gnb = GaussianNB()
# train with training data
gnb.fit(X_train, y_train)
# predict the result with testing data
y_pred_gnb = gnb.predict(X_test)
```

```python
dtc = tree.DecisionTreeClassifier()
# train with training data
dtc = dtc.fit(X_train, y_train)
# predict the result with testing data
y_pred_dt = dtc.predict(X_test)
```

```python
forest = RandomForestClassifier(n_estimators = 100)
# train with training data
forest = forest.fit(X_train, y_train)
# predict the result with testing data
y_pred_rf = forest.predict(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 2)
# train with training data
knn = knn.fit(X_train, y_train)
# predict the result with testing data
y_pred_knn = knn.predict(X_test)
```

```python
res_gnb = ["gaussian naive bayes", confusion_matrix(y_test, y_pred_gnb), accuracy_score(y_test, y_pred_gnb), recall_score(y_test, y_pred_gnb), precision_score(y_test, y_pred_gnb)]
res_dt = ["decision tree", confusion_matrix(y_test, y_pred_dt), accuracy_score(y_test, y_pred_dt), recall_score(y_test, y_pred_dt), precision_score(y_test, y_pred_dt)]
res_rf = ["random forest", confusion_matrix(y_test, y_pred_rf), accuracy_score(y_test, y_pred_rf), recall_score(y_test, y_pred_rf), precision_score(y_test, y_pred_rf)]
res_knn = ["KNeighbors", confusion_matrix(y_test, y_pred_knn), accuracy_score(y_test, y_pred_knn), recall_score(y_test, y_pred_knn), precision_score(y_test, y_pred_knn)]

output = pt.PrettyTable()
output.field_names = ["Method", "Confusion matrix", "Accuracy", "Recall", "Precision"]
output.add_row(res_gnb)
output.add_row(res_dt)
output.add_row(res_rf)
output.add_row(res_knn)
print(output)
```

```python
my_case = [[14, 2, 500, 26]]
my_case = np.array(my_case)
print(gnb.predict(my_case))
print(dtc.predict(my_case))
print(forest.predict(my_case))
print(knn.predict(my_case))
```
